#### Step1 Importing Libraries and IRIS Data

```
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
```

#### Seperating Features and Target

```
# X Features
X = pd.DataFrame(data=iris.data, columns=iris.feature_names)
X.head()
```

```
# y Target
y = iris.target
y
```

#### Step3: Fit decision tree regressor to the dataset


```
# import the regressor 
from sklearn.tree import DecisionTreeRegressor  
from sklearn.model_selection import train_test_split

# create a regressor object 
regressor = DecisionTreeRegressor(random_state = 0)  

# Splitting data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)

# fit the regressor with X and Y data 
regressor.fit(X_train, y_train)
print("Model is trained")

# evaluating score
accuracy = regressor.score(X_test, y_test)
print("Accuracy on test dataset is:", accuracy)
```

#### Visualize Decision Tree with Plot Tree

```
import matplotlib.pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(regressor, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

#### Visualize Decision Tree with graphviz

```
import graphviz

#DOT data
dot_data = tree.export_graphviz(regressor, out_file=None, 
                                feature_names=iris.feature_names,  
                                class_names=iris.target_names,
                                filled=True)

#Draw graph
graph = graphviz.Source(dot_data, format="png") 
graph
```
